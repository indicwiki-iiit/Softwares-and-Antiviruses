```python
def getData(row):
    data = {

        'Name':str(row['Name']).strip(),
        'Original_authors': str(row['Original_authors']).strip(), 
        'Developers': str(row['DEV']).strip(), 
        'Initial_release': str(row['Initial_release']).strip(),
        'Stable_version':str(row['Stable_version']).strip(),
        'Stable_version_release_date':str(row['Stable_version_release_date']).strip(),
        'preview_release':str(row['preview_release']).strip(),
        'written_in':str(row['written_in']).strip(),
         'repository':str(row['repository']).strip(),
        'middleware':str(row['middleware']).strip(),
        'engine':str(row['engine']).strip(),
        'os':str(row['os']).strip(),
        'platform':str(row['platform']).strip(),
        'Available_in':str(row['Available_in']).strip(),
        'type':str(row['type']).strip(),
        'License':str(row['License']).strip(),
        'as_of':str(row['as_of']).strip(),
        'website':str(row['website']).strip(),
        'included_with':str(row['included_with']).strip(),
        'service_name':str(row['service_name']).strip(),
        'size':str(row['size']).strip(),
        'description':str(row['description']).strip(),
        'related_softwares':str(row['related_softwares']).strip(),
        'additional_requirements':str(row['additional_requirements']).strip(),
        'whats_new':str(row['whats_new']).strip(),
        'references_':str(row['references_']).strip(),
        'external_Links':str(row['external_Links']).strip(),
        'category_':str(row['category_']).strip(),
        'image_files':str(row['image_files']).strip()
                              }                     
    return data;
```

```python
def main():
    with open("fullTemplate.j2",encoding="utf8") as f:
        S_template = Template(f.read())
    fobj = open("final_XML.xml", 'w',encoding="utf8")
    fobj.write(tewiki+'\n')

    for i in range(5):
        row = df.iloc[i]
        title = str(row['Name']).strip()
        text = S_template.render(getData(row))
        print(text)
        writePage(title, text, fobj)
        print('\n')
        #print('\n', i, title)
    fobj.write('</mediawiki>')
    fobj.close()
```

```python
main()
```
